I am using news dataset which contain articles of 'buisness' , 'sports'. We can use this same approach to classify text.

```python
import pandas as pd
```

```python
data = pd.read_csv(r"C:\Users\HIM\Desktop\Articles.csv", encoding = "ISO-8859-1")
```

```python
data
```

```python
import re
def cleaning(s):
    
    s = str(s)
    s = s.lower()
    s = re.sub('\s\W',' ',s)
    s = re.sub('\W,\s',' ',s)
    s = re.sub(r'[^\w,]', ' ', s)
    s = re.sub("\d+", "", s)
    s = re.sub('\s+',' ',s)
    s = re.sub('[!@#$_]', '', s)
    s = s.replace("co","")
    s = s.replace("https","")
    s = s.replace(",","")
    s = s.replace("[\w*"," ")
    return s
```

```python
data["Article"] = data["Article"].str.replace("strong>","")
data["clean"] = data["Article"].apply(cleaning)
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
```

```python
encoder = LabelEncoder()
y = encoder.fit_transform(data['NewsType']) # Converting categorial value for model
```

```python
count = CountVectorizer(min_df=2 , ngram_range=(1,3))
X = count.fit_transform(data["clean"])        
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier

# Instantiate the classifier: clf
clf = LogisticRegression()

# Fit the classifier to the training data
clf.fit(x_train, y_train)

# Print the accuracy
print("Accuracy: {}".format(clf.score(x_test, y_test )))
```

```python
a = clf.predict(x_test)
```

```python
from sklearn import metrics
fpr, tpr, thresholds = metrics.roc_curve(y_test,a)
print("AUC score: {} ".format(metrics.auc(fpr, tpr)))
```

```python
print(metrics.classification_report(y_test, a, target_names=encoder.classes_))
```
